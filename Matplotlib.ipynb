# Matplotlib

Matplotlib is a popular plotting library in Python that allows you to create various types of visualizations, such as line plots, bar charts, scatter plots, histograms, and more. It provides a wide range of customization options to create publication-quality figures.

Note: To run the code in Jupyter Notebook, make sure to include the `%matplotlib inline` magic command at the beginning of the notebook to display the plots inline.

```python
%matplotlib inline
```

Let's explore some examples to showcase the capabilities of Matplotlib:

### 1. Line Plot:

This code creates a simple line plot with data points `(1, 2), (2, 4), ..., (5, 10)` and adds labels and a title to the plot.

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')

# Display the plot
plt.show()
```

### 2. Bar Chart: 
This example generates a bar chart with four categories and their corresponding values.

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 25]

# Create a bar chart
plt.bar(categories, values)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')

# Display the plot
plt.show()
```

### 3. Scatter Plot:
In this example, a scatter plot is generated with random data points. The color and size of the points are determined by additional arrays, `colors` and `sizes`.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 100 * np.random.rand(50)

# Create a scatter plot
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')

# Display the plot
plt.show()
```

### 4. Histogram:
Here, a histogram is created using random data sampled from a normal distribution. The number of bins and edgecolor can be customized based on your preferences.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data from a normal distribution
np.random.seed(0)
data = np.random.randn(1000)

# Create a histogram
plt.hist(data, bins=30, edgecolor='black')

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')

# Display the plot
plt.show()
```

These examples demonstrate only a fraction of what Matplotlib can do. The library provides extensive options for customizing plots, such as adding legends, annotations, grids, and using different color maps.
